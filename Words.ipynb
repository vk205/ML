**Задачи**

 Выделить ключевые слова в тексте:
- Считать текст из файла. Рассчитать частоту слов в тексте. 
- Привести слова к нормальной форме с помощью библиотеки pymorphy2.
- С помощью межквартильных размахов выделить слова, которые являются выбросами в полученной выборке. 
- Являются ли выбросы стоп-словами

```python
import pymorphy2
```

#### Чтение текста из файла

```python
text = open('text_football_broadcast.txt', "r", encoding='utf-8-sig').read()
text
```

```python
import re

words = re.findall(r"[\w']+", text)
words
```

#### Частота слов

```python
from collections import Counter
```

```python
term_freq = Counter(words)

term_freq_values = [(key,value) for key,value in term_freq.items()]
sorted(term_freq_values, key=lambda x: x[1], reverse=True)
```

#### Приведение к нормальной форме

```python
morph = pymorphy2.MorphAnalyzer()

words = [morph.parse(word.lower())[0].normal_form
         for word in words]
```

```python
term_freq = Counter(words)
term_freq
```

```python
term_freq_values = [(key,value) for key,value in term_freq.items()]
sorted(term_freq_values, key=lambda x: x[1], reverse=True)
```

#### Выделение слов-выбросов 

```python
import numpy as np

Q1, Q2, Q3 = np.percentile(list(term_freq.values()), [25,50,75])
IQR = Q3 - Q1
```

```python
upper_inner = Q3 + 3 * IQR
upper_inner
```

```python
lower_inner = Q1 - 3 * IQR
lower_inner
```

```python
# выбросы
outlier_words = [(key,value) for key,value in term_freq.items() if (value <= lower_inner) | (value >= upper_inner)]
sorted(outlier_words, key=lambda x: x[1], reverse=True)
```

```python
# без выбросов
term_freq_values_IQR = [(key,value) for key,value in term_freq.items() if (value > lower_inner) & (value < upper_inner)]
sorted(term_freq_values_IQR, key=lambda x: x[1], reverse=True)
```

#### Поиск стоп-слов среди выбросов

```python
from nltk.corpus import stopwords

russian_stopwords = stopwords.words("russian")
```

```python
stop_words = [(key,value) for key,value in outlier_words  if key in russian_stopwords]

stop_words
```

```python
len(stop_words)
```

```python
len(stop_words)/len(outlier_words) *100
```

Таким образом среди слов-выбросов встречаются 25 стоп-слов, что состовляет около 30% от всех слов-выбросов.
